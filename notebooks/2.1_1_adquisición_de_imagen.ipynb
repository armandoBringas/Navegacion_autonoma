## Librer√≠as

```python
import cv2
from PIL import Image
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
%matplotlib inline
```

## Lectura de imagen y sus dimensiones

```python
img_path = '../../img'
img = cv2.imread(f'{img_path}/Google-self-driving-car-007.jpg')
```

```python
img.shape
print(f'altura = {int(img.shape[0])}, pixels')
print(f'anchura = {int(img.shape[1])}, pixels')
```

## Despligue de imagen con OpenCV en una ventana independiente

```python
cv2.imshow('Google self driving car', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Usando Matplotlib

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
rgb_img.shape
```

## Conversi√≥n de imagen en escala de grises

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow('Google self driving car in grayscale', gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
gray_img.shape
```

## Conversi√≥n en un arreglo bidemensional de Numpy

```python
img_arr = np.asarray(rgb_img)
img_arr.shape
plt.imshow(img_arr)
```

## Usando PIL

```python
img2 = Image.open(f'{img_path}/Google-self-driving-car-007.jpg')
```

## Convertir imagen en arreglo

```python
img2_arr = np.asarray(img2)
img2_arr.shape
plt.imshow(img2_arr)
```

## Crear una copia y mostrar en plano rojo

```python
img_red = img2_arr.copy()
# plano verde se hace cero
img_red[:,:,1] = 0
# plano azul se hace cero
img_red[:,:,2] = 0
plt.imshow(img_red)
```

## Conversi√≥n a espacio HSV

```python
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2.imshow('Google self driving car in HSV', hsv_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

## Visualizaci√≥n canal de Hue

```python
plt.imshow(hsv_img[:,:,0])
plt.title('Hue')
```

## Visualizaci√≥n canal de Saturation

```python
plt.imshow(hsv_img[:,:,1])
plt.title('Saturation')
```

## Visualizaci√≥n canal de Value

```python
plt.imshow(hsv_img[:,:,2])
plt.title('Value')
```
