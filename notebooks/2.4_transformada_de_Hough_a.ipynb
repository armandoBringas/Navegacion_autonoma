```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img_path = '../img'
img_bgr = cv2.imread(f'{img_path}/image_lane_c.jpg')
img = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
img_arr = np.asarray(img)
plt.imshow(img_arr, cmap='gray')
```

```python
# arbitrary ROI
vertices = np.array([[(20,540),(200,300),(650,300),(850,540)]], dtype=np.int32)
# image in black with the same dimensions of the original image in grayscale
roi_img = np.zeros_like(img)
# the ROI will be displayed in white over the black image
gaga = cv2.fillPoly(roi_img, vertices, 255)
plt.imshow(gaga, cmap='gray')
```
